**What is a percepton and a threshold logic unit (TLU)?**

A TLU is a unit that transforms input $x_1,...,x_n$ to a single output $h$ using a linear function $z$ and a step function $step(z)$ using corresponding weights $w_1,...,w_n$.


perceptron.png

A perceptron is a network architecture, where one or multiple TLUs are organized in a single layer and transform the same inputs using linear and step functions, as well as matrix algebra to simultaneously compute multiple outputs.

perceptron2.png

Linear functions $z(X)$ for two inputs $x1, x2$,corresponding weights $w1, w2, w3, w4$ and bias terms $b1,b2$ for two TLUs in a perceptron might look like: 

$ z(X)_1 = w1*x1+w2*x2+b1$
$ z(X)_2 = w3*x1+w4*x2+b2$

A step function $step(z(X))$ might look like for each TLU in the perceptron alike:

$ step(z(X)) = 0$ if $z < 0 $

$ step(z(X)) = 1$ if $z >=0 $

Now to calculate an example we just need to define the variables like:
$w1 = 0,2; w2=0,8; w3=0,4; w4=0,6$

$x1 = 4; x2=1$*

$b1 = -1,9; b2=0,6$*


By applying the above defined linear function we receive -0,3 and 2,6 for each TLU in the perceptron. Using the step function, we receive 0 and 1 for the TLUs.

**What is a deep neuronal network? What are hidden layers?**

TLUs can be arranged into multiple layers, where the output layer produces the final outputs, the input layers represent the inputs given. Any intermediate layer is considered a *hidden layer*.

The term deep neural network describes these multi-layer TLU arrangements. *Deep* though is a fuzzy part of the definition. In the 1990s, a two-layer network was considered deep. Nowadays, dozens of layers get arranged regularly.

**What means feedforward neural network (FNN).**

If the signal flows only in one direction (from input to output) within a neural network, it is considered a feedforward neural network.


**Try to explain how backpropagation works!**

1. Assign random weights and biases to the neural network.
2. Let an instance pass the network.
3. Compare the actual outputs with the expected outputs by calculating the total error.
4. Calculate the partial derivative of the total error with regards to each of the weights assigned.
5. Correct each of the weights by using the calculated derivative and a learning rate.
6. Repeat Steps 2 through 5 until the training set is passed a specified number of times. (*epochs*)

It must be noted that while step 2 can be done using singular instances, most of the times mini-batches of for example 32 instances are used. Overall, the training set is passed multiple times.

**Why do we need activation functions, wouldn't it be easier just using linear functions?**

In Step 4 of the above outlined process of backpropagation, derivatives are used to compute the impact of each weight on the total error. A linear function would only proliferate a constant as derivative, which would not be helpful for the algorithm to work.

```python
# 2: Calculate sum of distances between a tensor a and b
def mnist_distance(a,b): return F.mse_loss(a,b).sqrt()/b.shape[0]
```

```python
# 3: For each image, decide if it is a 1 or 7 and calculate accuracy

def is_1(x): 
  l2_one = mnist_distance(x,stacked_train_one_tensors)
  l2_seven = mnist_distance(x,stacked_train_seven_tensors)
  return l2_one.item() < l2_seven.item()

one_count = 0
seven_count = 0

for one in stacked_test_one_tensors:
  if is_1(one):
    one_count +=1

for seven in stacked_test_seven_tensors:
  if not is_1(seven):
    seven_count +=1

accuracy_one = one_count / stacked_test_one_tensors.shape[0]
accuracy_seven = seven_count / stacked_test_seven_tensors.shape[0]


accuracy_one,accuracy_seven,(accuracy_one+accuracy_seven)/2
```

# Stochastic Gradient Descent (SGD)

For this exercise I ask you to read the chapter Stochastic Gradient Descent (SGD) from the Google Colab 04_mnist_basics.ipynb in paralell. The chapter starts with a single TLU, compare p. 304 in "Hands on Machine Learning". Go through all 7 steps which are an easy example of how Stochastic Gradient Descent works.

Our goal is to train a single TLU, which can decide if one number is larger then the other one. Therefore we create 100 random pairs with pyTorch and create a target vector which is eather 1 or 0.


```python
x = torch.randn((100, 2))
y = torch.where(x[:,0] > x[:,1], 1.0, 0.0)
print(x)
print(y)
```

Your task is to create a function f that is a single TLU, meaning that it summarizes x with weights a, b, c:

$ax_0+bx_1+c$

In Addition we are using a *sigmoid()* function as step function.

$f = \text{sigmoid}(ax_0+bx_1+c)$

```python
def f(x, params):
    a,b,c = params
    return sigmoid(a*x[:,0]+b*x[:,1]+c)

print(f(x, [3,-2,1]))
```

In addition to our TLU function, we need a loss function. Your task is to implement a absolute difference loss function, $∑|x_i-y_i|$, which counts the number of wrong guesses.

```python
def mae(preds, targets): return (preds-targets.abs()).sum()
```

Try to train your single TLU with the absolute difference loss function, use the following code. Choose an appropriate step weight `lr` and try to explain what is happing in each line.

```python
# Initializing learning rate
lr = 1e-5

def apply_step(params, prn=True):
    # Generating predictions based on input array x, current param set and function defined in f
    preds = f(x, params)
    # Calculating loss function for predictions and target values
    loss = mae(preds, y)
    # Calculate gradient for current param set
    loss.backward()
    # Change param set according to learning rate and calculated gradients
    params.data -= lr * params.grad.data
    # Reset gradients for next iteration
    params.grad = None
    # print params and current loss
    if prn: print(params);print(loss.item())
    return loss.item()

# I've modified the code so that random parameters are set over and over again until
# a loss of <0.01 is reached.
while True:
  # Initializing parameters for prediction function to three random values
  params = torch.randn(3).requires_grad_()
  loss = None
  # repeat 50 epochs
  for i in range(50): 
    loss = apply_step(params)
  if abs(loss) < 0.01:
    break
```
